```python
def get_product(x: int, y:int) ->int:
    return x*y
get_product(x=2,y="6")
```

```python
from ensure import ensure_annotations

@ensure_annotations
def get_product1(x: int, y:int) ->int:
    return x*y
get_product1(x=2,y="6")
```
